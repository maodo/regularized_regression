```python
X = data.drop("sales",axis=1)
```

```python
X.head()
```

```python
y = data[["sales"]]
```

```python
y.head()
```

## Création du model linéaire non régularisé

### Division des données en trainning et test set avec 30% des données comme test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)
```

### Création du model de régression

```python
lin_model = LinearRegression()
lin_model.fit(X_train,y_train)
print('Training score: {}'.format(lin_model.score(X_train, y_train)))
print('Test score: {}'.format(lin_model.score(X_test, y_test)))

y_pred = lin_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)

print('RMSE: {}'.format(rmse))
```

Sur les données d'entrainement on a un taux de réussite de 90%
Sur les données de test on a un taux de réussite de 86%
On a une erreur de 2

## Création du model linéaire régularisé

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
ridge = Ridge()
lasso = Lasso()
```

```python

steps = [
    ('scalar', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2)),
    ('model', Ridge(alpha=10, fit_intercept=True))
]

ridge_pipe = Pipeline(steps)
ridge_pipe.fit(X_train, y_train)

print('Training Score: {}'.format(ridge_pipe.score(X_train, y_train)))
print('Test Score: {}'.format(ridge_pipe.score(X_test, y_test)))
```

```python
y_pred = ridge_pipe.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)

print('RMSE: {}'.format(rmse))
```
